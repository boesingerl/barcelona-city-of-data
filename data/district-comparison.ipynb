```python
import pandas as pd
```

```python
pop = pd.read_csv('population.csv')
lif = pd.read_csv('life_expectancy_withdistrict.csv')
ded = pd.read_csv('deaths.csv')
bir = pd.read_csv('births.csv')
une = pd.read_csv('unemployment.csv')
```

```python
birth_df = bir[bir['Year'] == 2017].groupby('District.Name').sum()['Number'].to_frame('births')
pop_df   = pop[pop['Year'] == 2017].groupby('District.Name').sum()['Number'].to_frame('pop')
death_df = ded[ded['Year'] == 2017].groupby('District.Name').sum()['Number'].to_frame('deaths')
unemp_df = une[(une['Year'] == 2017) & (une['Month'] == 'January')].groupby('District.Name').sum()['Number'].to_frame('unemployed')
life_df  = lif.groupby('District.Name').mean()['2010-2014'].to_frame('life_exp')
```

```python
valid_col = ['Ciutat Vella',
 'Eixample',
 'Gràcia',
 'Horta-Guinardó',
 'Les Corts',
 'Nou Barris',
 'Sant Andreu',
 'Sant Martí',
 'Sants-Montjuïc',
 'Sarrià-Sant Gervasi']

joint_df = birth_df.join(life_df).join(pop_df).join(death_df).join(unemp_df)
joint_df = joint_df.reset_index()
joint_df = joint_df[joint_df['District.Name'].isin(valid_col)].set_index('District.Name')

joint_df['unemployed_per'] = joint_df['unemployed'] / joint_df['pop']
joint_df['births_per']     = joint_df['births']    / joint_df['pop']
joint_df['deaths_per']     = joint_df['deaths']   / joint_df['pop']


joint_df = joint_df.drop(columns=['unemployed', 'births', 'deaths'])
#joint_df.loc['Mean'] = joint_df.mean()
joint_df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(joint_df))
scaled_df.columns = joint_df.columns
scaled_df.index = joint_df.index

scaled_df['unemployed_score'] = 1 - scaled_df['unemployed_per']
scaled_df['deaths_score'] = 1 - scaled_df['deaths_per']

scaled_df = scaled_df.drop(columns=['unemployed_per','deaths_per'])
scaled_df.loc['Mean'] = scaled_df.mean()
```

```python
scaled_df.to_csv('district_comp.csv')
```
